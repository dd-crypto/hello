```python
import numpy as np
import pandas as pd
from collections import Counter
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('googleplaystore.csv')
df
```

```python
Counter(df['Type'])
```

```python
type_null=df[df['Type'].isnull()].index.tolist()
print(type_null)
```

```python
df['Type'].iloc[type_null]='Free'
print(sum(df['Type'].isnull()))
print(Counter(df['Type']))
```

```python
df['Installs']=df['Installs'].apply(lambda x: (x.replace(',','')))
df['Installs']=df['Installs'].apply(lambda x: (x.replace('+','')))
df['Installs']=df['Installs'].apply(lambda x: (x.replace('Free','0')))
df['Installs']=pd.to_numeric(df['Installs'])
```

```python
Counter(df['Content Rating'])
```

```python
content_rating_null=df[df['Content Rating'].isnull()].index.tolist()
print(content_rating_null)
```

```python
df['Content Rating'].iloc[content_rating_null]='Everyone'
print(sum(df['Content Rating'].isnull()))
print(Counter(df['Content Rating']))
```

```python
Counter(df['Current Ver'])
```

```python
current_ver_null=df[df['Current Ver'].isnull()].index.tolist()
print(current_ver_null)
```

```python
df['Current Ver'].iloc[current_ver_null]='Varies with device'
print(sum(df['Current Ver'].isnull()))
Counter(df['Current Ver'])
```

```python
Counter(df['Android Ver'])
```

```python
android_ver_null=df[df['Android Ver'].isnull()].index.tolist()
print(android_ver_null)
```

```python
df.isnull().sum()
```

```python
df['Size']=df['Size'].apply(lambda x: x.replace('Varies with device','0'))
df['Size']=df['Size'].apply(lambda x: x.replace('1,000+','0'))

df['Size']
l1=[]
for i in df['Size']:
    if 'M' in i or 'm' in i:
        i=(float(i[:-1]))*1024*1024
        l1.append(i)
    elif 'K' in i or 'k' in i:
        i=(float(i[:-1]))*1024
        l1.append(i)
    else:
        l1.append(i)
df['Size']=l1
df['Size']=pd.to_numeric(df['Size'])
```
