```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Weekly-Sales.csv')
df
```

```python
df1=pd.read_csv('StoreDetails.csv')
df1
```

```python
df2=pd.read_csv('Region.csv')
df2
```

```python
df3=pd.merge(df,df1,how='inner',on='Store')
df3
```

```python
df4=pd.merge(df2,df3,how='inner',on='Store')
df4
```

```python
df4.head()
```

•	show the change in ice cream sales over the period 2010 to2012

```python
a=df['Date'].apply(lambda x: x[-4:])
b=(df.groupby(a)['Weekly_Sales'].count())
b.plot.bar()
plt.savefig('sales.pdf')
```

	•	Compare sales against the different store details (i.e. temperature and unemployment) to see what the relationship between them is (data from StoreDetails.csv).

```python
a=df4.groupby('Temperature')['Weekly_Sales'].sum()
a=a.head()
a.plot.bar()
```

```python
a=(df4.groupby('Region')['Weekly_Sales'].sum())
a.plot.bar()
```

•	Compare the trends of all stores in a given year.

```python

a=df4.groupby(df4['Date_y'].apply(lambda x: x[-4:]))['Weekly_Sales'].count()
a.plot.bar()
plt.savefig('stores.png')
```
